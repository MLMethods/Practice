```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.model_selection import train_test_split
```

## Dataset

```python
from sklearn import datasets
```

```python
# Load the digits dataset (8x8 images)
digits = datasets.load_digits()

IMAGE_INDX = 3

print("Features:\n", digits["images"][IMAGE_INDX])
print("Target value:", digits.target[IMAGE_INDX])

plt.imshow(digits.images[IMAGE_INDX])
plt.show()
```

```python
X, y = digits["data"], digits["target"]
X.shape, y.shape
```

```python
RANDOM_STATE = 100
```

```python
# Split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.3, 
    random_state=RANDOM_STATE
)
```

## Discriminative Model

```python
# Fit logistic regression model
model = LogisticRegression(solver="newton-cg", random_state=RANDOM_STATE)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
# Plot model parameters per class as 8x8 images
plt.figure(figsize=(10,5))
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.title(f"class = {model.classes_[i]}")
    plt.imshow(model.coef_[i].reshape(-1, 8)) #, vmin=-200, vmax=200)
    # plt.colorbar()
    plt.axis("off")
plt.tight_layout()
plt.show()
```

## Generative Model

```python
from sklearn.naive_bayes import GaussianNB
from scipy.stats import norm
```

```python
# Fit gaussian naive bayes model
model = GaussianNB()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
# Plot model means (theta_) per class as 8x8 images
plt.figure(figsize=(10,5))
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.title(f"class = {model.classes_[i]}")
    plt.imshow(model.theta_[i].reshape(-1,8)) #, vmin=-200, vmax=200)
    # plt.colorbar()
    plt.axis("off")
plt.tight_layout()
plt.show()
```

```python
# Generate digits from the model

CLASS_TARGET = 0

samples = norm.rvs(
    size=[10, X.shape[1]],
    loc=model.theta_[CLASS_TARGET], 
    # scale=np.full_like(model.var_[CLASS_TARGET], 1)
    scale=np.sqrt(model.var_[CLASS_TARGET])
).reshape(-1, 8, 8)

samples[samples > 16] = 16
samples[samples < 0] = 0

plt.figure(figsize=(10,5))
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(samples[i])
    plt.axis("off")
plt.tight_layout()
plt.show()
```
